```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
from geodatasets import get_path
```

```python
df = pd.DataFrame(
    {
        "City": ["Buenos Aires", "Brasilia", "Santiago", "Bogota", "Caracas", "East Lansing", "Amsterdam", "Sharjah", "Hong Kong", "Bejing", "Fribourg", "Davis", "Philadelphia", "Lubbock"],
        "Country": ["Argentina", "Brazil", "Chile", "Colombia", "Venezuela", "USA", "Netherlands", "UAE", "China", "China", "Switzerland", "USA", "USA", "USA"],
        "Latitude": [-34.58, -15.78, -33.45, 4.60, 10.48, 42.737652, 52.3676, 25.3570, 22.3193, 39.9042, 46.8065, 38.5449, 39.9526, 33.5846],
        "Longitude": [-58.66, -47.91, -70.66, -74.08, -66.86, -84.483788, 4.9041, 55.4035, 114.1694, 116.4074, 7.1620, -121.7405, -75.1652, -101.8456],
#aalborg
#Columbus
#Talahassee
#Utah
#Indiana
#Washington State
#
#texas
    
    }
)
```

```python
gdf = geopandas.GeoDataFrame(
    df, geometry=geopandas.points_from_xy(df.Longitude, df.Latitude), crs="EPSG:4326"
)
```

```python
world = geopandas.read_file(get_path("naturalearth.land"))

# We restrict to South America.
#ax = world.clip([1, 1, 1, 15]).plot(color="white", edgecolor="black")  #[-90, -55, -25, 15]
ax = world.plot(color="white", 
                edgecolor="black", 
                figsize=(15, 10))  #[-90, -55, -25, 15]

# We can now plot our ``GeoDataFrame``.
gdf.plot(ax=ax, color="red")

plt.show()
```
